```python
import numpy as np
import pandas as pd
data=pd.read_csv("data1.csv")
data
```

```python
#HANDLING MISSING VALUES 
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan, strategy='mean')
data[['Age','Salary']]=si.fit_transform(data[['Age','Salary']])
```

```python
data
```

```python
data['Country'].unique()
```

```python
#encoding categorical column (Onehot Encoding)

from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output =False)
encoded_data= ohe.fit_transform(data[['Country']])
encoded_data
```

```python
column_names=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encoded_data, columns=column_names)
encoded_df
```

```python
#joining old dataset to new dataset
data=pd.concat([data,encoded_df],axis=1)
data
```

```python
#droping categorial column 
data=data.drop(columns =['Country'])
data
```
