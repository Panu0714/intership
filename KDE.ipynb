```python
import matplotlib.pyplot as plt
import seaborn as sns

x=[1,2,3,4,5]
y=[28,33,64,23,67]

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("data1")
plt.ylabel("data2")
plt.show()

```

```python
import pandas as pd
data=pd.DataFrame({
    "class":['A','B','C','D','E'],
    "Students":[23,45,66,35,78]
})
sns.barplot(x="class",y="Students",data=data)
plt.title("Bar plot")
plt.xlabel("class")
plt.ylabel("Student")
plt.show()
```

```python
sns.countplot(data=data,x="class")
plt.title("Count plot")
plt.xlabel("class")
plt.ylabel("Students")
plt.show()
```

```python
import numpy as np

data=np.random.randn(100)
plt.hist(data,bins=10)
plt.title("Histogram")
plt.show()
```

```python
sns.kdeplot(data,shade=True)
plt.title("Kde plot")
plt.show()
```

```python
sns.boxplot(y=data)
plt.title("BOX plot")
plt.show()
```

```python
sns.violinplot(y=data)
plt.title("Violinplot")
plt.show()
```

```python
df=sns.load_dataset("iris")
df
```

```python
sns.pairplot(df,hue='species')
plt.title("pairplot")
plt.show()
```

```python
df_numeric=df.drop('species',axis=1)
co_relmat=df_numeric.corr()
co_relmat
```

```python
df_numeric=df.drop('species',axis=1)
co_relmat=df_numeric.corr()
co_relmat
sns.heatmap(co_relmat, annot=True,cmap='coolwarm')
plt.title("Heat map")
plt.show()
```

```python
labels=['A','B','C']
size=[30,50,20]

plt.pie(size,labels=labels,autopct='%1.1f%%',startangle=0)
plt.title("Pie plot")
plt.show()
```

```python
sns.swarmplot(x='species',y='sepal_length',data=df)

plt.title("Swarm plot")
plt.show()
```

```python
sns.stripplot(x='species',y='sepal_length',data=df,jitter=True)

plt.title("Strip plot")
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='scatter')

plt.title("joint plot")
plt.show()
```
